# Functional Method

```python
# Functional Method
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

# Object Oriented Method

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

# left, bottom, width, height
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.15,0.3,0.3,0.3])

axes1.plot(x,y)
axes1.set_title('Outer Plot')



axes2.plot(y,x)
axes2.set_title('Inner Plot')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3)

plt.tight_layout()

# for current_ax in axes:
#     current_ax.plot(x,y)

axes[0].plot(x,y)
axes[0].set_title('First Plot')
axes[0].set_xlabel('X Label')
axes[0].set_ylabel('Y Label')

axes[1].plot(y,x)
axes[2].plot(x,x)
```
